```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/aaronrao/Desktop/projects/Global Coding Challenge/day17.csv")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
charges = df["charges"]

print("Mean:", charges.mean())
print("Median:", charges.median())
print("Std Dev:", charges.std())
```

```python
sns.histplot(df["charges"], kde=True)
plt.show()

sns.scatterplot(x="age", y="charges", data=df)
plt.show()

sns.boxplot(x="smoker", y="charges", data=df)
plt.show()
plt.figure()
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```

## Key Insights

1. Smokers have significantly higher medical charges.
2. Charges increase with age.
3. Charges distribution is right-skewed.
4. BMI moderately affects charges.
5. Region has minimal impact.
